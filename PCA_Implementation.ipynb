```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
#Loading the dataset
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
## Standardization

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

```python
# Applying PCA Algorithm
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
data_pca=pca.fit_transform(scaled_data)
```

```python
data_pca
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
pca2 = PCA(n_components=3)
```

```python
data_pca2=pca2.fit_transform(scaled_data)
```

```python
data_pca2
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import Axes3D
import numpy as np
import pandas as pd
import seaborn as sns

# ... (your existing code) ...

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')  # Create a 3D subplot

# Scatter plot using the first 3 principal components
ax.scatter(data_pca2[:, 0], data_pca2[:, 1], data_pca2[:, 2], c=cancer['target'], cmap='plasma')

ax.set_xlabel('First Principal Component')  # Use ax.set_xlabel instead of plt.xlabel
ax.set_ylabel('Second Principal Component')  # Use ax.set_ylabel instead of plt.ylabel
ax.set_zlabel('Third Principal Component')  # Use ax.set_zlabel instead of plt.zlabel
ax.set_title('PCA with 3 Components')  # Use ax.set_title instead of plt.title

plt.show()
```
